```python
import qiskit
import numpy as np
from math import pi
from qiskit import QuantumCircuit
from qiskit import execute
from qiskit.quantum_info import Operator, Statevector, DensityMatrix, partial_trace
```

```python
circuit = qiskit.QuantumCircuit(2)
circuit.rx(pi, 0)
circuit.rz(pi / 2, 1)
circuit.cx(1, 0)
circuit.rz(pi / 6, 0)
circuit.rx(pi / 2, 1)
circuit.draw(style="mpl")
```

```python
unitary = Operator(circuit)

print(unitary)
```

```python
initial_state = Statevector([0.5, 0.5, 0.5, 0.5])
final_state = initial_state.evolve(circuit)
```

```python
print("Initial state:")
initial_state.draw(output="latex")
```

```python
print("Final state:")
final_state.draw(output="latex")
```

```python
initial_matrix = DensityMatrix(initial_state)
final_matrix = initial_matrix.evolve(circuit)

from qiskit.visualization import plot_state_city
plot_state_city(final_state)
```

```python
print("Initial matrix:")
initial_matrix.draw(output="latex")
```

```python
print("Final matrix:")
final_matrix.draw(output="latex")
```

```python
initial_0 = partial_trace(initial_matrix, [1])
initial_1 = partial_trace(initial_matrix, [0])

final_0 = partial_trace(final_matrix, [1])
final_1 = partial_trace(final_matrix, [0])
```

```python
print("Initial reduced density matrix for qubit 0:")
initial_0.draw(output="latex")
```

```python
print("Initial reduced density matrix for qubit 1:")
initial_1.draw(output="latex")
```

```python
print("Final reduced density matrix for qubit 0:")
final_0.draw(output="latex")
```

```python
print("Final reduced density matrix for qubit 1:")
final_1.draw(output="latex")
```
